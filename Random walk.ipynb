```python
#Random walk
from random import randint
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

walk = [100]
pos=100
for i in range(500):
    direction=randint(-1,1)
    pos = pos + direction
    walk.append(pos)
plt.plot(range(len(walk)),walk)

Variance = np.var(walk,ddof = 1)
print(Variance)
```
